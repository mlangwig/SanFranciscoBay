```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from matplotlib import pyplot as plt

```

##  test

```python
site4=pd.read_csv("Site4.tsv", sep="\t",index_col=0)
```

```python
site4.head()
```

```python
colorpalettte=sns.palplot(sns.color_palette("tab20b",19))
colors=sns.color_palette("tab20b",19)
```

```python
sns.set_style("white")
plt.figure(figsize=(7,5))
plt.xticks(size='medium')
plt.xticks(rotation=45)
#plt.yticks(size='xx-large')
plt.ylabel("Relative abundance", size="x-large", weight='bold')          
plt.axline(0.01,color='salmon', ls="--", linewidth=1) 
plt.axhspan(0, 0.01, alpha=0.1, color='salmon')
plt.ylim(0,50)
```

```python
sns.set_style("white")
plt.figure(figsize=(7,5))
```

```python
sns.set_style("white")
sns.set(font_scale=1)
ax1=site4.plot(kind='area',stacked=True,color=colors,figsize=(9,5))
plt.ylabel("Relative Abundance", weight='bold')
plt.xlabel("Time", weight="bold")
plt.title("Site 4.1", weight="bold", size="large")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.05 ,labelspacing=0.25)
lines, labels = ax1.get_legend_handles_labels()
ax1.legend(lines[:19], labels[:19], bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.05 ,labelspacing=0.3)
plt.tight_layout()
plt.savefig("site4.1.pdf")
```
